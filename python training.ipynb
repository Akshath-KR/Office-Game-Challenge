## Hangman Game

```python
HANGMANPICS = ['''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']
```

```python
import random
word_lists = ["Apple", "Orange", "Lemon", "Berries"]
HANGMANPICS.reverse()
lives = 6

random_word = random.choice(word_lists).lower()
print(random_word)

placeholder = ""
for num in range(0, len(random_word)):
    placeholder += "_"
print(placeholder)

game_over = False
correct_word = []

while not game_over: 
    guess = input("Guess a letter and I confirm if the letter is part of the word ").lower()
    display = ""
    for letter in random_word:
        if letter == guess:
            display += letter
            correct_word.append(guess)
        elif letter in correct_word:
            display += letter
        else:
            display += "_"

    print(display)

    if guess not in random_word:
        lives -= 1
        if lives == 0:
            game_over = True
            print("You Lose")

    if "_" not in display:
        game_over = True
        print("You Won")
    print(HANGMANPICS[lives])
    
```

```python
def calculate_love_score(name1, name2):
    check_1 = name1.lower() + name2.lower()
    sum_check_1 = 0
    for i in check_1:
        if i in "true":
            sum_check_1+= 1
            
    sum_check_2 = 0
    for j in check_1:
        if j in "love":
            sum_check_2+= 1
    
    sum_check_1 = str(sum_check_1)
    sum_check_2 = str(sum_check_2)
    
    print(sum_check_1+sum_check_2)
```

```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n").lower()
text = input("Type your message:\n").lower()
shift = int(input("Type your shift number:\n"))

def encrypt(orginal_text, shift_number):
    cipher_text = ""
    for letter in orginal_text:
        shifted_position = alphabet.index(letter) + shift_number
        shifted_position %= len(alphabet)
        cipher_text += alphabet[shifted_position]
    print(F"Here is the encoded result: {cipher_text}")    


```

```python
def decrypt(orginal_text, shift_number):
    output_text = ""
    for letter in orginal_text:
        shifted_position = alphabet.index(letter) - shift_number
        shifted_position %= len(alphabet)
        output_text += alphabet[shifted_position]
    print(F"Here is the descrypted result: {output_text}")    
```

```python
def ceser(orginal_text, shift_number, encode_or_decode):
    output_text = ""
    for letter in orginal_text:

        if encode_or_decode == "decode":
            shift_number *= -1
        
        shifted_position = alphabet.index(letter) - shift_number
        shifted_position %= len(alphabet)
        output_text += alphabet[shifted_position]
    print(F"Here is the {encode_or_decode}d result: {output_text}")   
```

```python
ceser(orginal_text = text, shift_number = shift, encode_or_decode = direction)
```

## Blind Action project

```python
bid_project = {}

new_bid = True

while new_bid is False:
    name = input("Your name?")
    bid = int(input("Your bid?: $"))
    bid_project[name] = bid

    new_bid_required = input("New bid required? Yes or No: ").lower()
    if new_bid_required == "no":
        new_bid = False
    else:
        print("\n" * 100)

max_name = max(bid_project)
max_value = bid_project[max_name]

print(f"Winner is {max_name} with the value of ${max_value}")
```
